# Covariance of random variables (and vectors)

See Section 3.7 in [SWJ]

$$
\operatorname{Cov}(X, Y)=\mathbb{E}\left[\left(X-\mu_{X}\right)\left(Y-\mu_{Y}\right)\right]=\mathbb{E}[X Y]-\mu_{x} \mu_{y}
$$

$$
\rho_{X Y}=\frac{\operatorname{Cov}(X, Y)}{\sqrt{\operatorname{Var}(X) \operatorname{Var}(Y)}}
$$

$$
\mu_{X}:=\left[\mathbb{E}\left[X_{1}\right], \mathbb{E}\left[X_{2}\right], \ldots, \mathbb{E}\left[X_{n}\right]\right]^{T}
$$

$$
\Sigma_{X}=\operatorname{Cov}(X)=\mathbb{E}\left[\left(X-\mu_{x}\right)\left(X-\mu_{x}\right)^{T}\right]
$$

$$
\mathbf{Y}=A \mathbf{X}+\mathbf{b}
$$

$$
\mathbb{E}[\mathbf{Y}]=A \mathbb{E}[\mathbf{X}]+\mathbf{b} \quad \text { and } \quad \operatorname{Cov}(\mathbf{Y})=A \Sigma_{\mathbf{X}} A^{T}
$$

Listing 3.32
https://github.com/h-Klok/StatsWithJuliaBook/blob/master/3_chapter/multiRVgeneration.jl

```julia
using Distributions, LinearAlgebra, LaTeXStrings, Random, Plots; pyplot()
Random.seed!(1)

N = 10^5

SigY = [ 6 4 ;
         4 9]
muY = [15 ; 
       20]
A = cholesky(SigY).L

rngGens = [()->rand(Normal()), 
           ()->rand(Uniform(-sqrt(3),sqrt(3))),
           ()->rand(Exponential())-1]

rv(rg) = A*[rg(),rg()] + muY
    
data = [[rv(r) for _ in 1:N] for r in rngGens]

stats(data) = begin
    data1, data2 = first.(data),last.(data)
    println(round(mean(data1),digits=2), "\t",round(mean(data2),digits=2),"\t",
            round(var(data1),digits=2), "\t", round(var(data2),digits=2), "\t",
            round(cov(data1,data2),digits=2))
end

println("Mean1\tMean2\tVar1\tVar2\tCov")
for d in data
    stats(d)
end

scatter(first.(data[1]), last.(data[1]), c=:blue, ms=1, msw=0, label="Normal")
scatter!(first.(data[2]), last.(data[2]), c=:red, ms=1, msw=0, label="Uniform")
scatter!(first.(data[3]),last.(data[3]),c=:green, ms=1,msw=0,label="Exponential",
	xlims=(0,40), ylims=(0,40), legend=:bottomright, ratio=:equal,
    xlabel=L"X_1", ylabel=L"X_2")
```

# Gaussians

$$
f(\mathrm{x})=(2 \pi)^{-n / 2} e^{-\frac{1}{2} \mathrm{x}^{T} \mathrm{x}}
$$

Listing 3.33
https://github.com/h-Klok/StatsWithJuliaBook/blob/master/3_chapter/multiDimensionalIntegration.jl

```julia
using HCubature

M = 4.5
maxD = 10

f(x) = (2*pi)^(-length(x)/2) * exp(-(1/2)*x'x)

for n in 1:maxD
    a = -M*ones(n)
    b = M*ones(n)
    I,e = hcubature(f, a, b, maxevals = 10^7)
    println("n = $(n), integral = $(I), error (estimate) = $(e)")
end
```

f(\mathrm{x})=\left|\Sigma_{\mathrm{X}}\right|^{-1 / 2}(2 \pi)^{-n / 2} e^{-\frac{1}{2}\left(\mathrm{x}-\mu_{\mathrm{X}}\right)^{T} \Sigma_{\mathrm{X}}^{-1}\left(\mathrm{x}-\mu_{\mathrm{X}}\right)}
$$

$$
f_{X Y}\left(x, y ; \sigma_{X}, \sigma_{Y}, \mu_{X}, \mu_{Y}, \rho\right)=\frac{1}{2 \pi \sigma_{X} \sigma_{Y} \sqrt{1-\rho^{2}}} 
~~ \exp \left\{\frac{-1}{2\left(1-\rho^{2}\right)}\left[\frac{\left(x-\mu_{X}\right)^{2}}{\sigma_{X}^{2}}-\frac{2 \rho\left(x-\mu_{X}\right)\left(y-\mu_{Y}\right)}{\sigma_{X} \sigma_{Y}}+\frac{\left(y-\mu_{Y}\right)^{2}}{\sigma_{Y}^{2}}\right]\right\}
$$

# The Sample Covariance Matrix

Section 4.2 in [SWJ]

$$
z_{i j}=\frac{x_{i j}-\bar{x}_{j}}{s_{j}}, \quad i=1, \ldots, n, \quad j=1, \ldots, p
$$

$$
Z=\left(X-1 \bar{x}^{T}\right) \operatorname{diag}(s)^{-1}
$$

$$
\hat{\rho}_{j_{1} j_{2}}=\frac{\sum_{i=1}^{n} z_{i j_{1}} z_{i j_{2}}}{n-1}=\left[\frac{1}{n-1} Z^{T} Z\right]_{j_{1} j_{2}}
$$


$$
\hat{\Sigma}=\frac{1}{n-1}\left(X-1 \bar{x}^{T}\right)^{T}\left(X-1 \bar{x}^{T}\right)=\frac{1}{n-1} X^{T}\left(I-n^{-1} 11^{T}\right) X
$$
