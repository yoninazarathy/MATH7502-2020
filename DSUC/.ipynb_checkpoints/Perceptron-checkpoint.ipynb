### The perceptron algorithm 

See 
https://www.cs.cornell.edu/courses/cs4780/2018fa/lectures/lecturenote03.html
for a good description.

```julia
using Random, Distributions, Plots; pyplot()

M = 2

function dataGenerator(n = 100)
    Random.seed!(0)
    w = [1,1]
    b = 0.5
    xPts = [[rand(Uniform(-M,M)), rand(Uniform(-M,M))] for _ in 1:n]
    label(x) = w'*x + b > 0 ? +1 : -1
    y = label.(xPts)
    return xPts, y
end
```

```julia
x, y = dataGenerator(200);
xNeg = x[y .== -1]
xPos = x[y .== +1]
scatter(first.(xNeg),last.(xNeg),c = :blue, ms = 5, msw=0, label = "-1 label")
scatter!(first.(xPos),last.(xPos),c = :red, ms = 5, msw=0, label = "+1 label",legend = :topright)
```

```julia
function learnPerceptron(xx,yy)
    n = length(xx)
    w = zeros(3)
    i = 1
    history = []
    while true
        misses = 0
        for i in 1:n
            predY = w'*[1 ; x[i]] > 0 ? +1 : -1 
            if yy[i] != predY
                push!(history,(w,i))
                w = w + yy[i]*[1 ; xx[i]]
                misses += 1
            end
        end
        if misses == 0
            break
        end
    end
    return w, history
end
```

```julia
extendedW ,history = learnPerceptron(x,y)
b = extendedW[1]
w = extendedW[2:end]
@show b
@show w
@show length(history)
```

```julia
scatter(first.(xNeg),last.(xNeg),c = :blue, ms = 5, msw=0, label = "-1 label")
scatter!(first.(xPos),last.(xPos),c = :red, ms = 5, msw=0, label = "+1 label")
x2Low = (-b - w[1]*(-M))/w[2]
x2High = (-b - w[1]*M)/w[2]
plot!([-M,M],[x2Low,x2High],lw = 5, c=:black,xlim=(-M,M),ylim=(-M,M),label ="Threshold")
```

```julia
pyplot()
anim = Animation()
for (n,h) in enumerate(vcat(history,fill(history[end],20)))
    scatter(first.(xNeg),last.(xNeg),c = :blue, ms = 5, msw =0, label = "-1 label")
    scatter!(first.(xPos),last.(xPos),c = :red, ms = 5, msw =0, label = "+1 label")
    i = h[2]
    b = h[1][1]
    w = h[1][2:end]
    x2Low = (-b - w[1]*(-M))/w[2]
    x2High = (-b - w[1]*M)/w[2]
    scatter!([first(x[i])],[last(x[i])], color = :green, ms = 10,label = "Last point")
    plot!([-M,M],[x2Low,x2High],lw = 5, c=:black,xlim=(-M,M),ylim=(-M,M),label ="Threshold")
    annotate!(0, -1.5, text("Iteration $(min(n,length(history))) out of $(length(history))", :center, :top, 20))
    frame(anim)
end
cd(@__DIR__)
gif(anim,"perceptron.gif",fps=5)
```
